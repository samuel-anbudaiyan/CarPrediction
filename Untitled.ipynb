```python
df_wo_Car_Name = df.drop(columns="Car_Name")
```

```python
df_Final = pd.get_dummies(df_wo_Year,drop_first=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = df_Final["Selling_Price"]
X = df_Final.drop(["Selling_Price"],axis = 1)
```

```python
y
```

```python
#Feature Importances
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
print(model.feature_importances_)
```

```python
series_feature_importance = pd.Series(model.feature_importances_,X.columns)
```

```python
series_feature_importance.plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
rf = RandomForestRegressor()
```

```python

#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]

param_dict = {"n_estimators" : n_estimators,
              "max_depth" : max_depth,
              "min_samples_split" : min_samples_split,
              "min_samples_leaf" : min_samples_leaf,
              "max_features" : max_features}
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold, StratifiedKFold
RandomCV = RandomizedSearchCV(rf,param_dict,scoring="neg_mean_squared_error",cv=5)
RandomCV.fit(X_train,y_train)
```

```python
RandomCV.best_params_
```

```python
RandomCV.best_score_
```

```python
pred = RandomCV.predict(X_test)
```

```python
sns.distplot(pred)
```

```python
sns.scatterplot(y_test,pred)
```
